# Data Quality Audit Project
E-commerce Dataset Analysis

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ecommerce_orders.csv')
df.head()
```

```
# Missing Values
missing = df.isnull().sum()
missing.plot(kind='bar', title='Missing Values')
plt.show()
```

```
# Consistency
df['calc_total'] = df['quantity'] * df['price']
df[df['total_amount'] != df['calc_total']]
```

```
# Accuracy
df[df['price'] <= 0]
df[df['quantity'] <= 0]
```

```
# Timeliness
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df[df['ship_date'] < df['order_date']]
```
